**Topic: Medical Diagnosis System based on Fuzzy Logic**
>   In this notebook, a Medical Diagnosis System is presented for predicting the level of Malaria infection. There are 11 input variables with 4 level: Minor, Moderate, Serve, and Very Serve. The output will be the conclusion also has 4 levels as mentioned.

**Reference source:** Awotunde, Joseph B., Opeyemi E. Matiluko, and Olawale W. Fatai. "Medical diagnosis system using fuzzy logic." African Journal of Computing & ICT 7.2 (2014): 99-106.
